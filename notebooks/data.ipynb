```python
import pandas as pd
import plotly.express as px
```

```python
sens = pd.read_csv("data/station_sensor_data.csv",parse_dates=["timestamp"])
sens['timestamp'] = sens['timestamp'].dt.round('min')
sens.head(10)
```

```python
sens['timestamp'].dt.round('min')[:10]
```

```python
sens['compressor_4_activity'].value_counts()
```

```python
sens.info()
```

```python
fig = px.histogram(
    sens, 
    x=['compressor_1_flow', 'compressor_2_flow', 'compressor_3_flow', 'compressor_4_flow','station_flow'  ], 
    barmode='overlay', 
    opacity=0.6,
    nbins=40,
    title="Comparing Column Distributions (Wide Format)"
)

# Customizing axes labels
fig.update_layout(
    xaxis_title="Values",
    yaxis_title="Count",
    legend_title="Columns"
)
fig.show()
```

```python
sens['specific_power_compressor_1'] = sens['compressor_1_power'] / sens['compressor_1_flow']
sens['specific_power_compressor_2'] = sens['compressor_2_power'] / sens['compressor_2_flow']
sens['specific_power_compressor_3'] = sens['compressor_3_power'] / sens['compressor_3_flow']        
sens['specific_power_compressor_4'] = sens['compressor_4_power'] / sens['compressor_4_flow']
fig = px.histogram(
    sens, 
    x=['specific_power_compressor_1', 'specific_power_compressor_2', 'specific_power_compressor_3', 'specific_power_compressor_4','station_specific_power'  ], 
    barmode='overlay', 
    opacity=0.6,
    nbins=50,
    title="Comparing Column Distributions (Wide Format)"
)
fig.show()
```
